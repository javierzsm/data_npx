```python
import pyopenephys
import matplotlib.pyplot as plt
import numpy as np
```

```python
data_path = r'Z:\raw\Javier Zorrilla\neurop_data\Kilosort_spike_sorting_results\Ts65Dn_145\Record Node 101\\'
```

```python
file = pyopenephys.File(data_path) 

# all experiments
experiments = file.experiments
print(len(experiments))

# recordings of first experiment
experiment = experiments[0]
recordings = experiment.recordings
print(len(experiments))

# access first recording
recording = recordings[0]

# experiment 1 (0 in Python)
experiment = file.experiments[0]
# recording 1 
recording = experiment.recordings[0]

print('Duration: ', recording.duration)
analog_signals = recording.analog_signals
```

```python

# select signals (low- or high-pass filtered)
signals = analog_signals[1]
```

```python
fs = signals.sample_rate
fs
```

#
Plots LFP channels
#
Substracts nearby channel to capture local voltage variations

```python
y_space = 100
t_plot = 2
ch_plot = np.arange(0, 384, 4)

fig_an, ax_an = plt.subplots(figsize=(10, 10))

for i in np.arange(0, 264, 1):
    time_plot = signals.times[:fs * t_plot]
    signal_plot = signals.signal[i][:fs * t_plot]+ i * y_space
    ax_an.plot(time_plot, signal_plot, linewidth=0.5)
ax_an.set_axis_off()
plt.show()
```

#
Plots LFP channels
#
Substracts nearby channel to capture local voltage variations

```python

ch_diff = 20
y_space = 100
t_plot = 2

fig_an, ax_an = plt.subplots(figsize=(10, 10))

for i in np.arange(0, 264, 1):
    st = (signals.signal[i][:fs * t_plot] - signals.signal[i + ch_diff][:fs * t_plot]) + i * y_space
    ax_an.plot(signals.times[:fs * t_plot], st, linewidth=0.5)
ax_an.set_axis_off()
plt.show()
```
